```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df1['reviewText'] = df1['reviewText'].apply(remove_stopwords)
```

```python
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk

# Ensure NLTK tokenizers are available
nltk.download('punkt')



# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Define a function to apply stemming to a given text
def stem_text(text):
    tokens = word_tokenize(text)  # Tokenize the text
    stemmed_words = [stemmer.stem(word) for word in tokens]  # Stem each word
    return ' '.join(stemmed_words)  # Rejoin stemmed words into a string

# Apply the stemming function to the 'Review' column
df1['reviewText'] = df1['reviewText'].apply(stem_text)

# Display the DataFrame after stem
```
